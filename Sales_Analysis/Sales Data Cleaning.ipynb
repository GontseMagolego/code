```python
import pandas as pd
import os
```

```python
files = [file for file in os.listdir('./Sales_Data')]

for file in files:
    print(file)
```

```python
data_for_year = pd.DataFrame()

for file in files:
    df = pd.read_csv('./Sales_Data/'+file)
    data_for_year = pd.concat([data_for_year, df])
    
data_for_year.to_csv('yearly_data.csv', index=False)
```

```python
yearly_data = pd.read_csv('yearly_data.csv')
yearly_data.head()
```

```python
nan_df = yearly_data[yearly_data.isna().any(axis=1)]
nan_df.head()
```

```python
yearly_data = yearly_data.dropna(how='all')
yearly_data.head()
```

```python
temp_df = yearly_data[yearly_data['Order Date'].str[0:2] == 'Or']
temp_df.head()
```

```python
yearly_data = yearly_data[yearly_data['Order Date'].str[0:2] != 'Or']
yearly_data.head()
```

```python
yearly_data['Quantity Ordered'] = pd.to_numeric(yearly_data['Quantity Ordered'])
yearly_data['Price Each'] = pd.to_numeric(yearly_data['Price Each'])
```

```python
yearly_data['Month'] = yearly_data['Order Date'].str[0:2]
yearly_data['Month'] = yearly_data['Month'].astype('int32')
yearly_data.head()
```

```python
yearly_data['Sales'] = yearly_data['Quantity Ordered'] * yearly_data['Price Each']
yearly_data.head()
```

```python
yearly_data.tail()
```

```python
null_counts = yearly_data.isnull().sum()
print(null_counts)
```

```python
year_data = yearly_data.groupby('Month').sum()
```

```python
df = pd.DataFrame(year_data)

df.to_excel('year_data.xlsx', index=False)
```
